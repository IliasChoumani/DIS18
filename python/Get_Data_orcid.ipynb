## Dict mit personen und deren Ids

```python
import requests

# Liste der ORCID-IDs
orcid_ids = [
    "0000-0002-1481-2996", "0000-0003-4561-9184", "0000-0002-2627-833X",
    "0000-0002-7229-5595", "0000-0001-5130-546X", "0000-0002-7086-2568",
    "0000-0002-6382-7211", "0000-0002-9489-3182"
]

# Funktion zur Abfrage der ORCID-API
def get_orcid_info(orcid_id):
    url = f"https://pub.orcid.org/v3.0/{orcid_id}"
    headers = {"Accept": "application/json"}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Fehler bei der Abfrage der ORCID-ID {orcid_id}: Statuscode {response.status_code}")
        return None

# Dictionary zur Speicherung der ORCID-Informationen
orcid_info_dict = {}

# Informationen für jede ORCID-ID abrufen und im Dictionary speichern
for orcid_id in orcid_ids:
    info = get_orcid_info(orcid_id)
    if info:
        name = info.get("person", {}).get("name", {})
        given_name = name.get("given-names", {}).get("value", "N/A")
        family_name = name.get("family-name", {}).get("value", "N/A")
        full_name = f"{given_name} {family_name}"
        orcid_info_dict[orcid_id] = full_name

# Ausgabe des Dictionaries
print(orcid_info_dict)
```

## Orcid Seiten als json speichern

```python
import json

# Funktion zur Abfrage der ORCID-API und Speichern als JSON
def save_orcid_info(orcid_id, name):
    url = f"https://pub.orcid.org/v3.0/{orcid_id}"
    headers = {"Accept": "application/json"}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        # JSON-Daten extrahieren
        data = response.json()
        # Dateiname für die JSON-Datei festlegen (Leerzeichen durch Unterstriche ersetzen)
        filename = f"{name.replace(' ', '_')}.json"
        # JSON-Daten in eine Datei schreiben
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
        print(f"Die Daten für die ORCID-ID {orcid_id} wurden als {filename} gespeichert.")
    else:
        print(f"Fehler bei der Abfrage der ORCID-ID {orcid_id}: Statuscode {response.status_code}")

# Informationen für jede ORCID-ID abrufen und als JSON-Dateien speichern
for orcid_id, name in orcid_info_dict.items():
    save_orcid_info(orcid_id, name)
```

## Arbeitspläte Herausfinden

```python
import json

# Liste der JSON-Dateien
json_list = ["Konrad_U._Förstner.json", "Alexander_Goesmann.json", "Anke_Becker.json", "Barbara_Götz.json", "Franziska_Hufsky.json"
            , "Marius_Dieckmann.json", "Thomas_Clavel.json", "Peer_Bork.json"]

# Durchlaufen aller JSON-Dateien in der Liste
for json_file in json_list:
    # Öffnen und Laden der JSON-Datei
    with open(json_file, "r", encoding="utf-8") as file:
        data = json.load(file)

    # Extrahieren der Titel aller Beschäftigungen
    employment_titles = []
    for employment_group in data.get("activities-summary", {}).get("employments", {}).get("affiliation-group", []):
        for employment_summary in employment_group.get("summaries", []):
            title = employment_summary.get("employment-summary", {}).get("organization", {}).get("name")
            if title:
                employment_titles.append(title)

    # Ausgabe der extrahierten Titel der Beschäftigungen für jede Datei
    print(f"Titel der Beschäftigungen in {json_file}:")
    for title in employment_titles:
        print(title)
```
